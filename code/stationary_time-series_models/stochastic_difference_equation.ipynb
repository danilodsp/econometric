# Stochastic Difference Equation Models

Simple example, suppose that Federal Reserve's money supply is given by

$m_{t} = 1.03 m_{t-1}$

Suppose now, that the FED cannot perfectly control the money supply by $\rho$ percentage

$\Delta m_{t} = \rho [m_{t} - m{t-1}] + \epsilon_{t}$ ,

we obtain

$m_{t} = \rho (1.03)^{t} m_{0} + (1-\rho)m_{t-1} + \epsilon_{t}$

where $\epsilon$ is the uncontrollable portion of the money supply.

```python
import matplotlib.pyplot as pp
```

```python
m_init = 0
t = range(100)
e = np.random.rand(len(t))
rho = 0.9
m = np.zeros(len(t))
```

```python
m_tmp = [rho*(1.03)*m_init + (1-rho)*m[i-1] + e[i] for i in range(1,len(t))]
m = np.concatenate(([0],m_tmp))
```

```python
pp.figure(figsize=(10,4))
pp.plot(t,m)
pp.xlabel('time (year)')
pp.ylabel('money supply')
```

Expectation of m

$E[m_{t+1}] = \rho*(1.03)^{t+1}*m_{0} + (1-\rho)*m_{t}$

```python
m_f = np.zeros(len(t))
m_f = [rho*((1.03)**(i))*m_init + (1-rho)*m[i-1] if i > 0 else 0 for i in range(0,len(t))]
```

```python
pp.figure(figsize=(10,4))
pp.plot(t,m,'b')
pp.plot(t,m_f,'r')
pp.xlabel('time (year)')
pp.ylabel('money supply')
```
